## Phân tích khu vực ảnh hưởng đến doanh thu

```python
# Chuyển về list 
df['production_countries'] = df['production_countries'].apply(ast.literal_eval)
```

```python
# Lấy top 12 khu vực có doanh thu cao nhất
revenue_by_area = {}
n_movie = {}
area_values = df['production_countries']
revenue = df['revenue']
for index, values in enumerate(area_values):
    if revenue[index] != 0:
        for value in values:
            if value['iso_3166_1'] in revenue_by_area:
                revenue_by_area[value['iso_3166_1']] += revenue[index]
                n_movie[value['iso_3166_1']] += 1
            else:
                revenue_by_area[value['iso_3166_1']] = revenue[index]
                n_movie[value['iso_3166_1']] = 1
                
d = collections.Counter(revenue_by_area)
top_revenue_by_area = []
for k, v in d.most_common(12):
    top_revenue_by_area.append(k)

print(top_revenue_by_area)
```

```python
# 
areas = []
revenues = [] 
area_values = df['production_countries']
revenue = df['revenue']
for index, values in enumerate(area_values):
    if revenue[index] != 0:
        for value in values:
            if value['iso_3166_1'] in top_revenue_by_area:
                revenues.append(revenue[index])
                areas.append(value['iso_3166_1'])

for key, value in revenue_by_area.items():
    if value != 0:
        avg_revenue_by_area[key] = float(value/n_movie[key])
l = [areas, revenues]
df2 = pd.DataFrame(l, index=['area','revenue']).T
df2.head()
```

```python

x = 'area'
y = 'revenue'

plt.figure(figsize=(18,6))
ax = sns.boxplot(x=x,y=y, data=df2)
plt.title('Revenue by Area')
plt.show()
```

Dựa vào biểu đồ ta có thấy các khu vực NZ có doanh thu trung bình cao nhất.

Để kết luận khu vực NZ có ảnh hưởng nhiều nhất đến doanh thu phim không, ta thực hiện Paired Samples t-test giữa 2 nhóm: doanh thu trên khu vực NZ và doanh thu trên những khu vực khác. 

Paired Sample t-test yêu cầu kích thước của 2 mẫu bằng nhau. Do đó, trước tiên tôi lấy toàn bộ doanh thu các bộ phim do khu vực NZ sản xuất tạo thành sample_1, sau đó tôi lấy ngẫu nhiên doanh thu của các bộ phim do các khu vực khác sản xuất tạo thành sample_2 sao cho kích thước sample_2 bằng kích thước sample_1

```python
sample_1 = df2[df2['area']=='NZ'].filter(['revenue'])
sample_2 = (df2[df2['area']!='NZ'].filter(['revenue'])).sample(sample_1.shape[0])
stats.ttest_rel(sample_2, sample_1)
```

Như vậy ta phản bác giả thuyết Ho và chấp nhận giả thuyết H1: Khu vực có ảnh hưởng đến doanh thu. Do đó khu vực NZ ảnh hưởng lớn nhất đến doanh thu phim 

## Ảnh hưởng của thể loại phim đến doanh thu và điểm của phim

```python
# Chuyển về list 
df['genres'] = df['genres'].apply(ast.literal_eval)
```

```python
genres = []
revenues = []
scores = []

genres_column = df['genres']
revenue_column = df['revenue']
score_column = df['vote_average']

for index, values in enumerate(genres_column):
    if revenue_column[index] != 0:
        for value in values:
            genres.append(value['name'])
            revenues.append(revenue_column[index])
            scores.append(score_column[index])          

l = [genres,revenues, scores]
df3 = pd.DataFrame(l, index=['genres','revenue', 'score']).T
df3.head()
```

```python
x = 'genres'
y = 'revenue'

plt.figure(figsize=(18,6))
ax = sns.boxplot(x=x,y=y, data=df3)
plt.title('Revenue by Genres')
plt.show()
```

Có thể thấy là doanh thu trung bình của thể loại Animation, Adventure và Fantasy cao nhất, còn những bộ phim thuộc thể loại Documentary, Foreign, có xu hướng kiếm được ít doanh thu. 

Ta có thể kiểm định để xem doanh thu có bị ảnh hưởng bởi thể loại phim hay không tương tự như câu trên

```python
sample_1 = df3[df3['genres']=='Animation'].filter(['revenue'])
sample_2 = (df3[df3['genres']!='Animation'].filter(['revenue'])).sample(sample_1.shape[0])
stats.ttest_rel(sample_1, sample_2)
```

```python
Như vậy ta phản bác giả thuyết Ho và chấp nhận giả thuyết H1: Thể loại có ảnh hưởng đến doanh thu. 
```

```python
x = 'genres'
y = 'score'

plt.figure(figsize=(18,6))
ax = sns.boxplot(x=x,y=y, data=df3)
plt.title('Vote average by Genres')
plt.show()
```

Đối với điểm của bộ phim thì thể loại Documentary có điểm trung bình cao nhất.
