## Phn tch khu vc nh hng n doanh thu

```python
# Chuyn v list 
df['production_countries'] = df['production_countries'].apply(ast.literal_eval)
```

```python
# Ly top 12 khu vc c doanh thu cao nht
revenue_by_area = {}
n_movie = {}
area_values = df['production_countries']
revenue = df['revenue']
for index, values in enumerate(area_values):
    if revenue[index] != 0:
        for value in values:
            if value['iso_3166_1'] in revenue_by_area:
                revenue_by_area[value['iso_3166_1']] += revenue[index]
                n_movie[value['iso_3166_1']] += 1
            else:
                revenue_by_area[value['iso_3166_1']] = revenue[index]
                n_movie[value['iso_3166_1']] = 1
                
d = collections.Counter(revenue_by_area)
top_revenue_by_area = []
for k, v in d.most_common(12):
    top_revenue_by_area.append(k)

print(top_revenue_by_area)
```

```python
# 
areas = []
revenues = [] 
area_values = df['production_countries']
revenue = df['revenue']
for index, values in enumerate(area_values):
    if revenue[index] != 0:
        for value in values:
            if value['iso_3166_1'] in top_revenue_by_area:
                revenues.append(revenue[index])
                areas.append(value['iso_3166_1'])

for key, value in revenue_by_area.items():
    if value != 0:
        avg_revenue_by_area[key] = float(value/n_movie[key])
l = [areas, revenues]
df2 = pd.DataFrame(l, index=['area','revenue']).T
df2.head()
```

```python

x = 'area'
y = 'revenue'

plt.figure(figsize=(18,6))
ax = sns.boxplot(x=x,y=y, data=df2)
plt.title('Revenue by Area')
plt.show()
```

Da vo biu  ta c thy cc khu vc NZ c doanh thu trung bnh cao nht.

 kt lun khu vc NZ c nh hng nhiu nht n doanh thu phim khng, ta thc hin Paired Samples t-test gia 2 nhm: doanh thu trn khu vc NZ v doanh thu trn nhng khu vc khc. 

Paired Sample t-test yu cu kch thc ca 2 mu bng nhau. Do , trc tin ti ly ton b doanh thu cc b phim do khu vc NZ sn xut to thnh sample_1, sau  ti ly ngu nhin doanh thu ca cc b phim do cc khu vc khc sn xut to thnh sample_2 sao cho kch thc sample_2 bng kch thc sample_1

```python
sample_1 = df2[df2['area']=='NZ'].filter(['revenue'])
sample_2 = (df2[df2['area']!='NZ'].filter(['revenue'])).sample(sample_1.shape[0])
stats.ttest_rel(sample_2, sample_1)
```

Nh vy ta phn bc gi thuyt Ho v chp nhn gi thuyt H1: Khu vc c nh hng n doanh thu. Do  khu vc NZ nh hng ln nht n doanh thu phim 

## nh hng ca th loi phim n doanh thu v im ca phim

```python
# Chuyn v list 
df['genres'] = df['genres'].apply(ast.literal_eval)
```

```python
genres = []
revenues = []
scores = []

genres_column = df['genres']
revenue_column = df['revenue']
score_column = df['vote_average']

for index, values in enumerate(genres_column):
    if revenue_column[index] != 0:
        for value in values:
            genres.append(value['name'])
            revenues.append(revenue_column[index])
            scores.append(score_column[index])          

l = [genres,revenues, scores]
df3 = pd.DataFrame(l, index=['genres','revenue', 'score']).T
df3.head()
```

```python
x = 'genres'
y = 'revenue'

plt.figure(figsize=(18,6))
ax = sns.boxplot(x=x,y=y, data=df3)
plt.title('Revenue by Genres')
plt.show()
```

C th thy l doanh thu trung bnh ca th loi Animation, Adventure v Fantasy cao nht, cn nhng b phim thuc th loi Documentary, Foreign, c xu hng kim c t doanh thu. 

Ta c th kim nh  xem doanh thu c b nh hng bi th loi phim hay khng tng t nh cu trn

```python
sample_1 = df3[df3['genres']=='Animation'].filter(['revenue'])
sample_2 = (df3[df3['genres']!='Animation'].filter(['revenue'])).sample(sample_1.shape[0])
stats.ttest_rel(sample_1, sample_2)
```

```python
Nh vy ta phn bc gi thuyt Ho v chp nhn gi thuyt H1: Th loi c nh hng n doanh thu. 
```

```python
x = 'genres'
y = 'score'

plt.figure(figsize=(18,6))
ax = sns.boxplot(x=x,y=y, data=df3)
plt.title('Vote average by Genres')
plt.show()
```
